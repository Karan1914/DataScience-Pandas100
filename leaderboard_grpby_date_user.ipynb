### Weekly Total Steps by per user 

```python
g1 = new_data.groupby([pd.Grouper(freq='W'),'user'])
result_week = g1['steps'].sum().unstack('user').fillna(0)
```

```python
result_week  # Weekly total steps by per user
```

### Monthly Total step count by per user 

```python
g1 = new_data.groupby([pd.Grouper(freq='M'),'user'])
result_mon = g1['steps'].sum().unstack('user').fillna(0)
result_mon 
```
